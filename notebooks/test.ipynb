```python
img = images[0]
bboxes = boxes[0]

plt.figure(figsize=(8,8))
plt.imshow(img)

# Draw bounding boxes
for box in bboxes:
    x1, y1, x2, y2, cls = box
    plt.gca().add_patch(plt.Rectangle((x1, y1), x2-x1, y2-y1,
                                      edgecolor='red', facecolor='none', linewidth=2))
    plt.text(x1, y1-5, str(cls), color='red', fontsize=10)

plt.axis('off')
plt.title("Preprocessed Image with Bounding Boxes")
plt.show()
```
